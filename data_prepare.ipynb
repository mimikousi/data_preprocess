```python
import pandas as pd
import numpy as np
```

```python
#データ読込み
df = pd.read_csv('debutanizer_time_series.csv', index_col=0, header=0)
```

```python
df
```

```python
#index(時間)をdatetime型に変換
df.index = pd.to_datetime(df.index)
```

```python
# 各列の標準偏差を計算
std_devs = df.std()
# 標準偏差が０のタグ数
print((std_devs == 0).sum())
```

```python
# 標準偏差が0でない列を選択
df_filtered = df.loc[:, std_devs != 0]
# 重複しているindexを削除（結合したデータの場合よくある）
df_filtered = df_filtered.loc[~df_filtered.index.duplicated(), :]
df_filtered
```

```python
# nanが多い列を削除
# 各列のNaNの割合を計算
nan_ratio = df_filtered.isna().mean()

# NaNの割合が0.2未満の列を選択
df_filtered_nan = df_filtered.loc[:, nan_ratio < 0.2]
df_filtered_nan
```

```python
# Xとyに分割する
X = df_filtered_nan.iloc[:, :-1]
y = df_filtered_nan.iloc[:, -1]
```

```python
X
```

```python
# 2〜10分の移動平均を各カラムに対して計算
ma_range = 10
X_ma = pd.DataFrame()
for col in X.columns:
    for window in range(2, ma_range+1):
        X_ma[f'{col}_ma_{window}'] = X[col].rolling(window=window).mean()
X_ma
```

```python
# 1〜30分のラグ特徴量を追加
delay_range = 30
lag_features = {}
for col in X.columns:
    for lag in range(1, delay_range+1):
        lag_features[f'{col}_lag_{lag}'] = X[col].shift(lag)
X_lag = pd.DataFrame(lag_features)
X_lag
```

```python
#データセットの結合
X_new = pd.concat([X, X_ma, X_lag], axis=1)
X_new = X_new.dropna()
X_new
```

```python
df_new = pd.concat([X_new, y], axis=1)
df_new = df_new.dropna()
df_new
```

```python
#特徴量が多くなりすぎるので、yとの相関が大きい変数のみ追加
# データフレームを用意
correlations_ma = pd.DataFrame()
correlations_lag = pd.DataFrame()
ma_range =40
delay_range = 30
```

```python
# 2〜40分の移動平均を各カラムに対して計算
for col in X.columns:
    for ma in range(2, ma_range+1):
        X_ma_ = X[col].rolling(ma).mean()
        corr = X_ma_.corr(y)
        correlations_ma.loc[ma, col] = corr
```

```python
# ラグ0から30までの特徴量を作成し、yとの相関を計算
for col in X.columns:
    for lag in range(delay_range + 1):
        X_lagged = X[col].shift(lag)
        corr = X_lagged.corr(y)
        correlations_lag.loc[lag, col] = corr
```

```python
# 各カラムに対して相関係数の絶対値が最大のラグを抽出
best_mas = correlations_ma.abs().idxmax()
```

```python
# 各カラムに対して相関係数の絶対値が最大のラグを抽出
best_lags = correlations_lag.abs().idxmax()
```

```python
# 最も相関が強いラグ特徴量だけを含むデータセットを作成
X_lag_cor = pd.DataFrame()
for col, lag in best_lags.items():
    if np.isnan(lag):
        X_lag_cor[f'{col}_lag_{0}'] = X[col]
    else:
        lag = int(lag)
        X_lag_cor[f'{col}_lag_{lag}'] = X[col].shift(lag)
```
